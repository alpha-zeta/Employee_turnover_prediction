```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression as logi
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.svm import SVC 
from sklearn.feature_selection import RFE 
```

```python
df=pd.read_csv("employee_data.csv")
cols=df.columns.to_list()
cols
```

```python
df.isnull().any()
```

```python
df["department"].unique()
```

```python
df.loc[df["department"]=="IT","department"]="technical"
df.loc[df["department"]=="support","department"]="technical"
df
```

```python
deps=["department","salary"]
for i in deps:
    lst=pd.get_dummies(df[i],prefix=i)
    df=df.join(lst)
```

```python
df.drop(df.columns[[8,9]],axis=1,inplace=True)
df.columns
```

```python
nclos=df.columns.to_list()
y=df["quit"]
x=df[[i for i in nclos if i!="quit"]]
```

```python
x
```

```python
model=logi()
rfe=RFE(model,10)
rfe=rfe.fit(x,y)
rank=rfe.ranking_
```

```python
nclos.remove("quit")
```

```python
ncols=[]
for i in range(len(rank)):
    if rank[i]==1:
        ncols.append(nclos[i])
ncols
x=df[ncols]
x
```

```python
xtr,xte,ytr,yte=tts(x,y,test_size=0.2)
xtr
```

```python
model=logi()
model.fit(xtr,ytr)
accuracy_score(yte,model.predict(xte))
```

```python
import joblib as jb
jb.dump(model,"emp_logistic")
```

```python
model=rfc()
model.fit(xtr,ytr)
accuracy_score(yte,model.predict(xte))
```

```python
import joblib as jb
jb.dump(model,"emp_rfc")
```

```python
model=SVC()
model.fit(xtr,ytr)
accuracy_score(yte,model.predict(xte))
```

```python
import joblib as jb
jb.dump(model,"emp_svc")
```
